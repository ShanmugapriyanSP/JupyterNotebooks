```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import time
from datetime import datetime
import seaborn as sns
```

```python
x_train = pd.read_csv('files/UCI HAR Dataset/UCI HAR Dataset/train/X_train.txt',delimiter=r"\s+", header=None)
y_train = pd.read_csv('files/UCI HAR Dataset/UCI HAR Dataset/train/y_train.txt', delimiter=r"\s+",header=None)
x_test = pd.read_csv('files/UCI HAR Dataset/UCI HAR Dataset/test/X_test.txt', delimiter=r"\s+",header=None)
y_test = pd.read_csv('files/UCI HAR Dataset/UCI HAR Dataset/test/y_test.txt', delimiter=r"\s+",header=None)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
# x_train = x_train.iloc[:,:].values
# x_test = x_test.iloc[:,:].values
# y_train = y_train.iloc[:,:].values
# y_test = y_test.iloc[:,:].values
x_train= np.asarray(x_train) 
y_train= np.asarray(y_train)
x_test= np.asarray(x_test) 
y_test= np.asarray(y_test)
```

```python
# from tensorflow.keras.models import Sequential
# from tensorflow.keras.layers import Dense

# # Initialising the ANN
# classifier = Sequential()

# # Adding the input layer and the first hidden layer
# classifier.add(Dense(units = 280, kernel_initializer = 'uniform', activation = 'relu', input_dim = 561))

# # Adding the second hidden layer
# classifier.add(Dense(units = 280, kernel_initializer = 'uniform', activation = 'relu'))

# classifier.add(Dense(units = 280, kernel_initializer = 'uniform', activation = 'relu'))

# # Adding the output layer
# classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))

# # Compiling the ANN
# classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# # Fitting the ANN to the Training set
# start = datetime.now()
# classifier.fit(x_train, y_train, batch_size = 10, epochs = 100)
# end = datetime.now()
# print(end - start)

from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
from sklearn import metrics
pscore = metrics.accuracy_score(y_test, y_pred)
```

```python
pscore
```

```python
from sklearn.model_selection import GridSearchCV
#help(GridSearchCV)
parameters = [{'n_estimators':[100,250,500,700], 'criterion' : ['gini'],
                'max_features':['auto', 'log2'], 'bootstrap':['True','False'],
                 'warm_start':['True','False']},
             {'n_estimators':[100,250,500,700], 'criterion' : ['entropy'],
             'max_features':['auto', 'log2'], 'bootstrap':['True','False'],
                 'warm_start':['True','False']}]
grid_search = GridSearchCV(estimator = classifier,
                          param_grid = parameters,
                          scoring = 'accuracy',
                          cv = 10,
                          n_jobs=-1)
grid_search = grid_search.fit(x_train, y_train)
print(grid_search.best_score_)
print(grid_search.best_params_)
print(grid_search.best_estimator_)
```
