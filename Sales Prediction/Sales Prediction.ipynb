# Artificial Neural Network

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
import tensorflow as tf
tf.test.is_gpu_available()
```

```python
x = x.reshape(-1,2)
x_tr = x_tr.reshape(-1,2)
x_te = x_te.reshape(-1,2)
y = y.reshape(-1,1)
y_tr = y_tr.reshape(-1,1)
y_te = y_te.reshape(-1,1)
```

```python
x_tr
```

```python
# from sklearn.preprocessing import StandardScaler
# sc_x = StandardScaler()
# sc_y = StandardScaler()
from sklearn.preprocessing import MinMaxScaler
sc_x = MinMaxScaler(feature_range=(0,1))
sc_y = MinMaxScaler(feature_range=(0,1))
```

```python
x = sc_x.fit_transform(x)
x_tr = sc_x.transform(x_tr)
x_te = sc_x.transform(x_te)
y = sc_y.fit_transform(y)
y_tr = sc_y.transform(y_tr)
y_te = sc_y.transform(y_te)
```

```python
y
```

```python
ann_regressor = Sequential()
```

```python
ann_regressor.add(Dense(units = 32, activation = 'relu', kernel_initializer = 'normal', input_dim = 2))
```

```python
ann_regressor.add(Dense(units = 64, activation = 'relu', kernel_initializer = 'normal'))
ann_regressor.add(Dense(units = 64, activation = 'relu', kernel_initializer = 'normal'))
ann_regressor.add(Dense(units = 64, activation = 'relu', kernel_initializer = 'normal'))
```

```python
ann_regressor.add(Dense(units = 1, kernel_initializer = 'normal', activation = 'linear'))
```

```python
ann_regressor.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_squared_error'])
```

```python
ann_regressor.summary()
```

```python
start = datetime.now()
ann_regressor.fit(x_tr, y_tr, batch_size=10, epochs=1)
end = datetime.now()
print(end - start)
```
